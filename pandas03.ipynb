```python
#pythonとデータフレームを接続する(ここではsqliteと接続)
import sqlite3
```

```python
df_csv = pd.read_csv('data.csv',encoding = 'shift-jis')
df_csv
```

```python
#データベースの作成
dbname = 'TEST.db'
#コネクションオブジェクト作成（よくわからん）
conn = sqlite3.connect(dbname)
#データベースの接続を解除
conn.close()
```

```python
#データフレームをデータベースのテーブルに作成
#if_existsは現在テーブルがある場合どのように操作するのかを決める
#append...テーブルに追加　　　replace...テーブルを削除してから新規で作成  fail...エラーを発生させる
conn = sqlite3.connect(dbname)
df_csv.to_sql('test_table',conn,if_exists = 'replace')
conn.close
```

```python
#実際作成されたかどうかデータを読み込んで確認してみる
conn = sqlite3.connect(dbname)
#SELECT * FROM test_tableはSQL文
df_db = pd.read_sql('SELECT * FROM test_table', conn)
conn.close()
```

```python
df_db
```

```python
#SQL文を以下に記述して東京都のみを抽出
sql = '''
select *
from test_table
where 都道府県名 == "東京都"
'''
```

```python
#上のデータをread_sqlに渡す
conn = sqlite3.connect(dbname)
df_db = pd.read_sql(sql, conn)
conn.close()
```

```python
df_db
```

```python
#大量データの扱い方
conn = sqlite3.connect(dbname)

for i in range(1000):
    df_csv.to_sql('test_table_1000',conn,if_exists = 'append')

conn.close()
#このコードを行うと1000*100=100万行のデータができるが、パソコンに負荷がかかるので実行しません！
```

```python
#SQL文で東京都の中で1920年のデータのみ抽出する
sql = '''
select *
from test_table_1000
where 都道府県名 = "東京都" and "西暦（年）" = "1920.0"
'''
```

```python
conn = sqlite3.connect(dbname)
df_db = pd.read_sql(sql, conn)
conn.close()
```
